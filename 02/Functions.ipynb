```python
def expdecay(xdata=np.linspace(0,.1,100),T2=0.025):
    R2=-1/T2
    signal= np.exp(xdata*R2)
    return signal
```

```python
# define variables
TR=np.linspace(0,.1,50);
myT2=.050;
```

```python
# estimate signal using both inputs
S1=expdecay(TR,myT2)
# estimate signal using only TR
S2=expdecay(TR)
# estimate signal using only T2
S3=expdecay(T2=myT2)
```

```python
# plot S1 to S3
plt.subplot(1,3,1); plt.plot(TR,S1,'-') 
plt.subplot(1,3,2); plt.plot(TR,S2,'--r') 
plt.subplot(1,3,3); plt.plot(S3,'sk') 
```

```python
# plot assuming both inputs
plt.plot(expdecay(),'-r') 
```

```python
# plot changing order of inputs
plt.plot(expdecay(T2=.030,xdata=TR),'-r')
plt.plot(expdecay(xdata=TR,T2=.030),'ob')
```
